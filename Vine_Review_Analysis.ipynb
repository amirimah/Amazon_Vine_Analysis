```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("vine_table.csv")
```

```python
helpful_df= df[df["total_votes"]>= 20]
helpful_df
```

```python
vine_df = helpful_df[(helpful_df["helpful_votes"]/helpful_df["total_votes"]>=0.5) & (helpful_df["vine"]=="Y")]
vine_df
```

```python
not_vine_df = helpful_df[(helpful_df["helpful_votes"]/helpful_df["total_votes"]>=0.5) & (helpful_df["vine"]=="N")]
not_vine_df
```

### total number of reviews, the number of 5-star reviews, and the percentage 5-star reviews

```python
five_star_Y=vine_df[vine_df['star_rating']==5]
five_star_Y = five_star_Y['review_id'].count()
five_star_Y
```

```python
five_star_N=not_vine_df[not_vine_df['star_rating']==5]
five_star_N= five_star_N['review_id'].count()
five_star_N
```

```python
total_reviews_Y = vine_df['review_id'].count()
total_reviews_Y
```

```python
total_reviews_N = not_vine_df['review_id'].count()
total_reviews_N
```

```python
percentage_five_star_Y = (five_star_Y/total_reviews_Y)*100
percentage_five_star_Y
```

```python
percentage_five_star_N = (five_star_N/total_reviews_N)*100
percentage_five_star_N
```
